```python
from langchain.chains import RetrievalQA
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import LlamaCpp
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.document_loaders import PyPDFDirectoryLoader
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#load pdf files
loader = PyPDFDirectoryLoader("/content/drive/MyDrive/ApolloMission_testdoc/")
data = loader.load()
```

```python
print(data)
```

```python
#Split the Extracted Data into Text Chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20)

text_chunks = text_splitter.split_documents(data)
```

```python
text_chunks[5]
```

```python
#Downlaod the Embeddings
embeddings = HuggingFaceEmbeddings(model_name="thenlper/gte-large")
```

```python
#Create Embeddings for each of the Text Chunk
vector_store = FAISS.from_documents(text_chunks, embedding=embeddings)
```

```python
#Import Model
llm = LlamaCpp(
    streaming = True,
    model_path="/content/drive/MyDrive/ApolloDocs/zephyr-7b-beta.Q4_K_M.gguf",
    temperature=0.75,
    top_p=1,
    verbose=True,
    n_ctx=4096
)
```

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vector_store.as_retriever(search_kwargs={"k": 2}))
```

```python
query = "Can you elaborate on the search strategy employed, including the databases searched, and the criteria used for article selection"
```

```python
qa.run(query)
```

```python
import pandas as pd

# List of queries
queries = ["Can you provide insights into the results of the systematic review ,including the number of studies identified, their characteristics, and any key findings?"

,"How were patient and public involvement considered in the research, and what impact did it have on the study?"]

# Initialize an empty list to store results
results = []

# Iterate through the queries
for query in queries:
    # Assuming llm and vector_store are already defined
    qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vector_store.as_retriever(search_kwargs={"k": 2}))

    # Perform retrieval for the current query
    answer = qa.run(query)

    # Append the result to the list
    results.append({"Query": query, "Answer": answer})

# Convert the list of results to a DataFrame
result_df = pd.DataFrame(results)
```
